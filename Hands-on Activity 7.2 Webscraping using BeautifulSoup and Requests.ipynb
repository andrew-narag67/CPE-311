```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

local_driver_path = "./geckodriver.exe"
firefox_binary_path = r"C:\Program Files\Mozilla Firefox\firefox.exe"

options = Options()
options.binary_location = firefox_binary_path
options.page_load_strategy = "eager"

options.set_preference("general.useragent.override",
                       "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                       "AppleWebKit/537.36 (KHTML, like Gecko) "
                       "Chrome/120.0.0.0 Safari/537.36")

service = Service(executable_path=local_driver_path)

driver = webdriver.Firefox(service=service, options=options)
driver.set_page_load_timeout(30)

driver.get("https://www.metacritic.com/browse/game/?releaseYearMin=1958&releaseYearMax=2026&page=1")

print("Page Title:", driver.title)

WebDriverWait(driver, 20).until(
    EC.presence_of_element_located((By.TAG_NAME, "h3"))
)
```

```python
import requests
from bs4 import BeautifulSoup
import time

names = []
scores = []
dates = []

base_url = "https://www.metacritic.com/browse/game/?releaseYearMin=1958&releaseYearMax=2026&page={}"

headers = {
    "User-Agent": "Mozilla/5.0"
}

for page in range(1, 6):
    url = base_url.format(page)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    time.sleep(3)

    for card in soup.find_all("div", class_="c-finderProductCard"):
        title = card.find("h3", class_="c-finderProductCard_titleHeading")
        score = card.find("div", class_="c-siteReviewScore")
        release_date = card.find("div", class_="c-finderProductCard_meta")

        if title:
            names.append(title.text.strip())

        if score:
            scores.append(score.text.strip())

        if release_date:
            clean_date = release_date.text.split("•")[0].strip()
            dates.append(clean_date)

print(names)
print(scores)
print(dates)
```

```python
import pandas as pd
data = {
    "Names": names,
    "Scores": scores,
    "Dates": dates
}

df = pd.DataFrame(data)
df
```
