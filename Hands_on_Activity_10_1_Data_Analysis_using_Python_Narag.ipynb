```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from matplotlib.gridspec import GridSpec
import warnings
warnings.filterwarnings('ignore')

plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.titlesize': 13,
    'axes.titleweight': 'bold',
    'axes.labelsize': 11,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'figure.dpi': 150,
})
ASEAN_BLUE   = '#005B96'
ASEAN_TEAL   = '#17A589'
ASEAN_ORANGE = '#E67E22'
ASEAN_RED    = '#C0392B'
ASEAN_PURPLE = '#6C3483'
ASEAN_GRAY   = '#7F8C8D'

df = pd.read_csv('/content/drive/MyDrive/Foldah/data-5tiX4.csv', encoding='utf-8-sig')
df.rename(columns={'X.1': 'Year'}, inplace=True)
df['Year'] = df['Year'].astype(int)


df_clean = df.copy()
df_clean.ffill(inplace=True)

df_hist = df_clean[df_clean['Year'] <= 2021].copy()
df_proj = df_clean[df_clean['Year'] > 2021].copy()

numeric_cols = [c for c in df_clean.columns if c != 'Year']
desc = df_clean[numeric_cols].describe().round(3)

# Figure 1
fig, axes = plt.subplots(4, 2, figsize=(16, 20))
fig.suptitle('ASEAN Labour Market Indicators (2010–2026)\nTrend Analysis',
             fontsize=16, fontweight='bold', y=0.98)

colors = [ASEAN_BLUE, ASEAN_TEAL, ASEAN_RED, ASEAN_ORANGE,
          ASEAN_PURPLE, ASEAN_GRAY, '#2E86C1', '#1E8449']

for ax, col, color in zip(axes.flatten(), numeric_cols, colors):
    # Historical
    ax.plot(df_hist['Year'], df_hist[col], color=color, lw=2.5,
            marker='o', markersize=4, label='Historical')
    # Projected
    ax.plot(df_proj['Year'], df_proj[col], color=color, lw=2,
            linestyle='--', marker='s', markersize=4, alpha=0.7, label='Projected')
    # COVID
    ax.axvspan(2019.5, 2021.5, alpha=0.12, color='red', label='COVID-19 period')
    ax.set_title(col)
    ax.set_xlabel('Year')
    ax.set_ylabel('%')
    ax.legend(fontsize=7)
    ax.grid(axis='y', alpha=0.3)
"""
ANALYSIS
The analysis shows that labour market indicators in ASEAN are closely connected. Higher vulnerable employment usually means more people are poor, since informal jobs are less stable. Labour force participation and unemployment don’t strongly move together, as many people still find work in the informal sector. More jobs in services are linked to better, more stable employment. Youth unemployment also rises with overall unemployment, showing young people are more affected, especially during COVID-19. Overall, these results show that labour issues are linked, so solutions should address multiple areas at once.
"""
```
