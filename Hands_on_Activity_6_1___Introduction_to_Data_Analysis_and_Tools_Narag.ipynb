```python
#median
def calculate_median(salaries):
  sorted_list = sorted(salaries)

  n = len(sorted)
  mid_index = n//2
  if n % 2 == 1:
    median = sorted_list(mid_index)
  else:
    median = (sorted_list[mid_index-1] + sorted_list[mid_index])/2

  return median
print(median)
```

```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
print(salaries)
```

```python
#mode
from collections import Counter
counts = Counter(salaries)
most_common = counts.most_common(1)
mode = most_common[0][0]
max_count = max(counts.values())
modes = [item for item, count in counts.items() if count == max_count]
print(mode)
print(modes)
```

# Using the counter library we can use counts to get the most common repeated value in the list therefore getting the mode.

```python
#variance
avg = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1

x_mean_squared = []
for salary in salaries:
  temp = (salary - avg)**2
  x_mean_squared.append(temp)


summation = sum(x_mean_squared)
variance = summation/len_minus_one


print(variance)
```

# To get the variance we first get the mean squared then dividing it to our list.

```python
#standard deviation
Sd = variance ** 0.5
print(Sd)
```

# Using the outputted variance we can multiply it to the power of 0.5 to get the standard deviation.
